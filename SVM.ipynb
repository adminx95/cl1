 Implementation of Support Vector Machines (SVM) for classifying images of
hand- written digits into their respective numerical classes (0 to 9)

```python
from sklearn.datasets import load_digits
```

```python
x=load_digits()['data']
y=load_digits()['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
```

```python
svc.fit(x_train,y_train)
```

```python
svc.score(x_test,y_test)
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

i = 34
plt.imshow(x_train[i].reshape((8,8)));
plt.title(f"Actual: {y_train[i]} Predicted: {svc.predict([x_train[i]])[0]}");
```

```python
from sklearn.metrics import classification_report

y_preds = svc.predict(x_test)

print(classification_report(y_test,y_preds))
```
